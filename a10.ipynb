```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("iris.csv")
```

```python
df.sample(5)
```

```python
df.info()
```

```python
cols = ['sepal.length', 'sepal.width',
        'petal.length', 'petal.width']

plt.figure(figsize=(10,8))
plt.subplot(2, 2, 1)
sns.histplot(data=df, x=cols[0], kde=True)

plt.subplot(2, 2, 2)
sns.histplot(data=df, x=cols[1], kde=True)

plt.subplot(2, 2, 3)
sns.histplot(data=df, x=cols[2], kde=True)

plt.subplot(2, 2, 4)
sns.histplot(data=df, x=cols[3], kde=True)
```

```python
plt.figure(figsize=(10,8))
plt.subplot(2, 2, 1)
sns.boxplot(data=df, y=cols[0])

plt.subplot(2, 2, 2)
sns.boxplot(data=df, y=cols[1])

plt.subplot(2, 2, 3)
sns.boxplot(data=df, y=cols[2])

plt.subplot(2, 2, 4)
sns.boxplot(data=df, y=cols[3])
```

```python
q1=df['sepal.width'].quantile(0.25)
q3=df['sepal.width'].quantile(0.75)

iqr=q3-q1
lower=q1-1.5*iqr
upper=q3+1.5*iqr
```

```python
outliers = df[
    (df['sepal.width'] < lower) |
    (df['sepal.width'] > upper)
]
print(outliers)
```
