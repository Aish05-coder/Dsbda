```python
import pandas as pd

# Load dataset
df = pd.read_csv("adult.csv")

# Print first 5 rows
print(df.head())
```

```python
summary_stats = df.groupby('income')['hours.per.week'].describe()

print("Summary Statistics:")
print(summary_stats)
```

```python
median_hours = df.groupby('income')['hours.per.week'].median()

print("\nMedian Hours per Week by Income:")
print(median_hours)
```

```python
print("Column Names:")
print(df.columns)
```

```python
mode_age = df['age'].mode()
mode_hours = df['hours.per.week'].mode()

print("Mode of Age:", mode_age.values)
print("Mode of Hours per Week:", mode_hours.values)
```

```python
gender_stats = df.groupby('sex')['age'].describe()

print("\nAge Statistics by Gender:")
print(gender_stats)
```

```python
import pandas as pd

df = pd.read_csv("iris.csv")

print(df.head())
```

```python
summary_statistics = df.groupby('species').describe()

percentile = df.groupby('species').quantile([0.25,0.5,0.75])

print("Percentile of Species : ")
print(percentile)
```

```python
summary_stats_species = df.groupby('species')['sepal_width'].describe()

print("\nSummary Statistics for Sepal Width:")
print(summary_stats_species)
```

```python
median_values = df.groupby('species').median()

print("\nMedian Values by Species:")
print(median_values)
```

```python
median_sepal_length = df.groupby('species')['sepal_length'].median()

print("\nMedian Sepal Length by Species:")
print(median_sepal_length)
```

```python
mode_width = df['sepal_width'].mode()

print("Mode of Sepal Width:", mode_width.values)
```
